```python
import pandas as pd
import numpy as np
```

 ## Series = A pandas 1-dimensional labeled array that can hold any data type Think of it like a single column in a spreadsheet(1-dimensional)

```python
data = np.array([100, 102, 104])

series = pd.Series(data, index = ["a","b","c"])

print(series.loc['c']) # for accessing the index directly by calling
print(series.iloc[0])  # for accessing the index from index number for it
print("   ")

new_data = np.array([100,200,300,400,500,600])

new_series = pd.Series(new_data)
print(new_series[new_series<500])

print("   ")

calories = {
    'Day 1' : 1750,
    'Day 2' : 2100,
    'Day 3' : 1700
}

Dict_series = pd.Series(calories)

Dict_series['Day 3'] += 500  # Adds 500 in Day 3 and save new value
print(Dict_series.loc['Day 3'])
print("   ")
print(Dict_series)
print("   ")
print(Dict_series[Dict_series<2000])
```

##  Dataframe = A tabular data structure with rows and column. (2 Dimensional) Similar to an Excel spreadsheet

```python
another_data = {
    'Name': ["spongebob", "patrick", "squidward"],
    'Age': [30,35,50]
}

df = pd.DataFrame(another_data, index=["Employee 1", "Employee 2", "Employee 3"])
print(df)
print("  ")

print(df.loc["Employee 3"])
print("  ")

# Add a new column

df['Job'] = ["Cook", "N/A", "Cashier"]
print(df)
print("  ")

# Add a new row
new_row = pd.DataFrame([{
    'Name': "sandy",
    'Age': "28",
    'Job': "Engineer"    
},
                        {
    'Name': "Eugene",
    'Age': "60",
    'Job': "Manager"    
}], index=["Emplayee 4","Employee 5"])

df = pd.concat([df, new_row])
print(df)
```

## Filtering = keeping the rows that match a condition

```python
tall_pokemon = df[df["Height"]>=2]
# print(tall_pokemon)

weight_pokemon = df[df["Weight"]> 100]
# print(weight_pokemon)

water_fire_pokemon = df[(df["Type1"] == "Water") | (df["Type2"] == "Fire")]
print(water_fire_pokemon)
```

## aggregate functions = Reduces a set of values into a single summary value Used to summarize and analyze data often used with the groupby() function

```python
print(df.mean(numeric_only=True))
print(df.sum(numeric_only=True))
print(df.min(numeric_only=True))
print(df.max(numeric_only=True))
print(df.count())

# Single column

print(df["Height"].mean())
print(df["Height"].sum())
print(df["Height"].min())
print(df["Height"].max())
print(df["Height"].count())

group = df.groupby("Type1")
print(group["Height"].mean())
print(group["Height"].sum())
print(group["Height"].count())
```


## Data cleaning = the process of fix/removing: incomplete, incorrect, or irrelevant data.
