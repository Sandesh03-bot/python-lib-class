```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023,2024,2025,2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([3, 12, 9, 22])

line_style = dict(
     marker = ".",
     markersize = 20,
     markerfacecolor = "#1cd3fc",
     markeredgecolor =  "#1cd3fc",
     linestyle = "solid",
)

plt.title("Class size", fontsize = 20, color = "#fc1c1c")
plt.xlabel("Year", fontsize = 20)
plt.ylabel("students", fontsize = 20)

plt.xticks(x)


plt.plot(x,y1, **line_style) # passing list x and y1 as coordinates
plt.plot(x,y2, **line_style) # passing list x and y2 as coordinates
plt.plot(x,y3,color = "#fc721c", **line_style) # passing list x and y3 as coordinates
plt.show()
```

```python
# gird() = helps make plots easier to read by addin reference lines.

x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.grid(axis="y",
         linewidth = 2,
         color = "lightgray",
         linestyle = "dashed")
plt.plot(x,y)
plt.show()
```

```python
# Bar chart = Compare categories of data by representing each category with a bar

categories = ["Grains", "Fruit", "Vegetables", "Protein", "Dairy", "Sweets"]
values = np.array([4,3,2,5,3,1])

plt.title("Daily consumption")
plt.ylabel("Quantity", fontsize = 20)
plt.xlabel("Food", fontsize = 20)
plt.bar(categories,  values)
plt.show()
```

```python
# Pie chart = Circular chart divided into slices to show percentages of the total.

new_categories = ["Freshmen", "Sophomores","Juniors","Seniors"]
values = np.array([300,250,275,225])

colors = [
    "red",
    "yellow",
    "blue",
    "green"
]
plt.pie(values, labels=new_categories,
        autopct="%1.1f%%",
        colors=colors,
        explode=[0,0,0,0.1],
        shadow=True)

plt.show()
```

```python
# scatter graph = Shows the relationship betwwen two variables helps to identify a correlation(+, -, None)
                    #  Example: Study hours vs Test Scores
                    
x = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]
y1 = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]
y2 = [50, 62, 70, 22, 33, 35, 25, 78, 89, 58, 72]

plt.title("Test Scores")
plt.xlabel("Hours studied")
plt.ylabel("Grade")
plt.scatter(x, y1)
plt.scatter(x, y2)
plt.grid(axis="y")
plt.show()
```

```python
# Histogram = A viual representation of the distribution of quantitive data
                # They group Values into bins (intervals)
                # and counts how many fall in each range.
                
scores = np.random.normal(loc = 80, scale = 10, size = 100)
scores = np.clip(scores, 0, 100)
plt.hist(scores, bins = 10,color = "lightgreen",edgecolor = "black")
plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("No of Students")
plt.show()
```

```python
# Sub plot

# Figure = The entire canva
# Ax = A single plot (subplot)

x = np.array([1, 2, 3, 4, 5])

figure, axes =plt.subplots(2, 2)

axes[0, 0].plot(x, x*2, color = "red")
axes[0, 0].set_title("x*2")

axes[0,1].plot(x, x**2, color = "blue")
axes[0,1].set_title("x**2")

axes[1,0].plot(x, x**2, color = "green")
axes[1,0].set_title("x**3")

axes[1,1].plot(x, x**2, color = "yellow")
axes[1,1].set_title("x**4")

plt.tight_layout()

plt.show()
```

```python
# pandas + matplotlib

import pandas as pd

df = pd.read_csv("data.csv")

type_count = df["Type1"].value_counts(ascending=True)

plt.barh(type_count.index, type_count.values,edgecolor = "black")
plt.title("No of pokemon by primary type")
plt.xlabel("count")
plt.ylabel("Type")
plt.tight_layout
plt.show()
```
